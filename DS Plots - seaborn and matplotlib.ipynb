PairGrid Plot

```python
iris = sns.load_dataset('iris')
iris.head()

```

```python
x=sns.PairGrid(data=iris)
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_offdiag(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',vars=['petal_length','sepal_length'])
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(iris,hue='species',vars=['petal_length','petal_width'])
x=x.map_diag(plt.hist,edgecolor='black')
x=x.map_offdiag(plt.scatter,edgecolor='black')
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,x_vars=['petal_width'],y_vars=['petal_length'],hue='species')
x=x.map(plt.scatter,edgecolor='black')
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species')
x=x.map_diag(plt.hist)
x=x.map_upper(plt.scatter)
x=x.map_lower(sns.kdeplot)
x=x.add_legend()
```

```python
x=sns.PairGrid(iris,hue='species',hue_kws={'marker':['D','s','+']},size=4)
x=x.map(plt.scatter)
x=x.add_legend()
```

```python
#facetgrid
tips=sns.load_dataset('tips')
tips.head()
```

```python
x=sns.FacetGrid(data=tips,row='sex',col='time',hue='smoker',palette='winter_r',row_order=['Female','Male'])
x=x.map(plt.scatter,'total_bill','tip')
x=x.add_legend()
```

```python
x=sns.FacetGrid(tips,'smoker','time').map(sns.distplot,'tip',color='green',kde=False,bins=10)
```

```python
import matplotlib.pyplot as plt
sns.set_style(style='darkgrid')
x = sns.FacetGrid(data= tips, hue='day', size=5, row='smoker', col='time')
x= x.map(plt.scatter, 'total_bill', 'tip')
x= x.add_legend()
```

```python
x=sns.FacetGrid(tips,row='smoker',col='day',hue='sex').map(sns.regplot,'total_bill','tip').add_legend()
```
